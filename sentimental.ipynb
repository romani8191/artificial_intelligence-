```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
import requests
from bs4 import BeautifulSoup
import re
```

```python
# initiate model
```

```python
tokenizer= AutoTokenizer.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
model= AutoModelForSequenceClassification.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
```

```python
tokens= tokenizer.encode('Today was a lovely day . I met a really fun child in autorikshaw', return_tensors='pt')
```

```python
result=model(tokens)
```

```python
result.logits
```

```python
int(torch.argmax(result.logits))+1
```

```python
# collect reviews
```

```python
r= requests.get('https://www.yelp.com/biz/home-thai-restaurant-sydney-3?osq=Thai')
soup =BeautifulSoup(r.text,'html.parser')
regex=re.compile('.*comment.*')
results=soup.find_all('p',{'class':regex})
reviews=[result.text for result in results]
```

```python
# load reviews into dataframe and score
```

```python
import numpy as np
import pandas as pd
```

```python
df=pd.DataFrame(np.array(reviews), columns=['review'])
```

```python
df.head(4)
```

```python
df['review'].iloc[0]
```

```python
# creating the model to parse data and get sentiments
```

```python
def sentimental_analyser (review):
  tokens= tokenizer.encode(review, return_tensors='pt')
  result=model(tokens)
  return int(torch.argmax(result.logits))+1
```

```python
sentimental_analyser (df['review'].iloc[0])
```

```python
df['sentiment'] = df['review'].apply(lambda x: sentimental_analyser(x[:512]))
```
